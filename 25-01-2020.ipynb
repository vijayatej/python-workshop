```python
# adding grid
y=[1,2,3]
x=[1,2,3]
plt.title("graph")
plt.plot(x,y)
plt.grid()
```

```python
# adding lables
y=[1,2,3]
x=[1,2,3]
plt.title("graph")
plt.plot(x,y)
plt.xlabel("Amp--->")
plt.ylabel("time--->")
plt.grid()
```

```python
# adding dots
y=[1,2,3]
x=[1,2,3]
plt.title("graph")
plt.plot(x,y)
plt.xlabel("Amp--->")
plt.ylabel("time--->")
plt.grid()
plt.plot(x,y,"go")# r->red ,o->symbol to represent
```

```python
# ploting through list
x=list(range(1,10))
y=list(range(1,10))
plt.plot(x,y,"go")
plt.grid()
```
